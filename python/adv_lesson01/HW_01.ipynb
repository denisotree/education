Преобразовать слова «разработка», «администрирование», «protocol», «standard» из строкового представления в байтовое и выполнить обратное преобразование (используя методы encode и decode).

```python
str_11 = "разработка"
str_12 = "администрирование"
str_13 = "protocol"
str_14 = "standard"

bstr_11 = str_11.encode()
bstr_12 = str_12.encode()
bstr_13 = str_13.encode()
bstr_14 = str_14.encode()

dstr_11 = bstr_11.decode()
dstr_12 = bstr_12.decode()
dstr_13 = bstr_13.decode()
dstr_14 = bstr_14.decode()

print(bstr_11)
print(bstr_12)
print(bstr_13)
print(bstr_14)

print('\n')

print(dstr_11)
print(dstr_12)
print(dstr_13)
print(dstr_14)
```

Создать текстовый файл test_file.txt, заполнить его тремя строками: «сетевое программирование», «сокет», «декоратор». Проверить кодировку файла по умолчанию. Принудительно открыть файл в формате Unicode и вывести его содержимое.

```python
str_list = ['сетевое программирование', 'сокет', 'декоратор']

with open('test_file.txt', 'w') as f:
    f.write("\n".join(str_list))

with open('test_file.txt', 'r') as f:
    data = f.read()
    
print(data)

print('\n')

with open('test_file.txt', 'r', encoding="utf-8") as f:
    data_u = f.read()
    
print(data_u)
```
