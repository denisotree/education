```python
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
from pylab import meshgrid,linspace,zeros,dot,norm,cross,vstack,array,matrix,sqrt

def plane(Lx,Ly,Nx,Ny,n,d):

    x = linspace(-Lx/2,Lx/2,Nx)
    y = linspace(-Ly/2,Ly/2,Ny)
    X,Y = meshgrid(x,y)
    Z   = zeros([Nx,Ny])
    n0 = array([0,0,1])

    dVec = (n/norm(n))*d
    X,Y,Z = X+dVec[0],Y+dVec[1],Z+dVec[2]
    return X,Y,Z

distList = linspace(-5, 5, 2)

normalVector = array([0,0,1])

fig = plt.figure()
ax = fig.gca(projection='3d')

for i,ypos in enumerate(linspace(-10,10,2)):

    X,Y,Z = plane(20,20,100,100,normalVector,distList[i])

    ax.plot_surface(X, Y, Z, rstride=5, cstride=5, alpha=0.8)

 
ax.set_xlabel('X')
ax.set_xlim(-10, 10)
ax.set_ylabel('Y')
ax.set_ylim(-10, 10)
ax.set_zlabel('Z')
ax.set_zlim(-10, 10)

plt.show()
```

```python
fig = plt.figure(figsize=plt.figaspect(1))
ax = fig.add_subplot(111, projection='3d')

coefs = (4, 6, 6)
rx, ry, rz = 1/np.sqrt(coefs)

u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)

x = rx * np.outer(np.cos(u), np.sin(v))
y = ry * np.outer(np.sin(u), np.sin(v))
z = rz * np.outer(np.ones_like(u), np.cos(v))

ax.plot_surface(x, y, z,  rstride=6, cstride=6)

max_radius = max(rx, ry, rz)
for axis in 'xyz':
    getattr(ax, 'set_{}lim'.format(axis))((-max_radius, max_radius))
    
plt.show()
```

```python
par1 = [2, -5, 7]
par2 = [-1, 6, 0]
par3 = [1, 4, 1]
```

```python
x = linspace(-5, 5, 100)
```

```python
def y_val(x, k, a, b):
    y = []
    for i in x:
        y.append(k*np.cos(i - a) + b)
    return y
```

```python
plt.plot(x, y_val(x, *par1))
plt.plot(x, y_val(x, *par2))
plt.plot(x, y_val(x, *par3))
plt.axes().set_aspect(1)
```

#### Задание 3.3

```python
import numpy as np

def cart2pol(x, y):
    rho = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)
    return(rho, phi)

def pol2cart(rho, phi):
    x = rho * np.cos(phi)
    y = rho * np.sin(phi)
    return(x, y)
```
