```python
from math import factorial
from numpy import exp
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))

def bern(n, k, p):
    return combinations(n, k) * (p ** k) * ((1-p) ** (n - k))

def puasson(n, p, m):
    lambda_ = n * p
    return ((lambda_ ** m) / factorial(m)) * exp(-lambda_)
    
```

#### 1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

```python
P = bern(100, 85, 0.8)
print(f'Верояность попадания 85 раз {round(P, 5)}')
```

#### 2. Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?

```python
P1 = bern(5000, 0, 0.0004)
P2 = bern(5000, 2, 0.0004)
P3 = puasson(5000, 0.0004, 0)
P4 = puasson(5000, 0.0004, 2)
print(f'Вероятность того, что не перегороит ни одна лампочка {round(P1, 5)} по формуле Бернулли или {round(P3, 5)} по формуле Пуассона')
print(f'Вероятность того, что перегорит ровно две лампочки {round(P2, 5)} по формуле Бернулли или {round(P4, 5)} по формуле Пуассона')
```

#### 3. Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?

```python
P5 = bern(144, 70, 0.5)
print(f'Вероятность того, что орел выпадет ровно 70 раз {round(P5, 5)}')
```
