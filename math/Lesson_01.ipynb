```python
from math import factorial
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

#### 1. Из колоды в 52 карты извлекаются случайным образом 4 карты. Найти вероятность того, что все карты – крести. Найти вероятность, что среди этих карт окажется хотя бы один туз.

```python
n = combinations(52, 4)

print(f"Количество комбинаций как можно вытащить карты, {n}")
print(f"Количество карт одной масти 13, всего мастей 4")

nt = combinations(13, 4)

print(f"Количество комбинаций только для треф — {nt}")
print(f"Вероятность того, что все карты — трефы {nt/n*100}%")
print(f"Количество комбинаций для тузов в 4-ках карт — {combinations(4, 1)}")
print(f"Количество комбинаций для нетузов в 4-ках карт — {combinations(48, 3)}")
print(f"Следовательно, вероятность, что из 4х карт выпадет хотя бы один туз — {(combinations(48, 3)*4)/n*100}%")
```
